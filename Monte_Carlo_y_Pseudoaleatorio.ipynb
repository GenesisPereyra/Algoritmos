#Monte Carlo

```python
import numpy as np
import matplotlib.pyplot as plt

# Definimos la función f(x) = x^2
def f(x):
    return x ** 2

# Simulación de Monte Carlo
def monte_carlo_integration(num_samples):
    # Generamos números aleatorios uniformemente distribuidos en [0, 1]
    x_random = np.random.uniform(0, 1, num_samples)

    # Evaluamos la función en los puntos generados
    f_values = f(x_random)

    # Estimamos la integral como el promedio de los valores de f(x)
    integral_estimate = np.mean(f_values)

    return integral_estimate

# Número de muestras a generar
num_samples = 100000

# Estimamos la integral
estimated_integral = monte_carlo_integration(num_samples)

# Mostramos el resultado
print(f"Valor estimado de la integral usando Monte Carlo: {estimated_integral}")
print(f"Valor exacto de la integral: {1/3}")

# Gráfico de los puntos generados
x_random = np.random.uniform(0, 1, num_samples)
f_values = f(x_random)

plt.scatter(x_random, f_values, color="blue", alpha=0.1, label="Puntos aleatorios")
plt.plot(x_random, f_values, color="red", label="f(x) = x^2")
plt.title("Simulación de Monte Carlo - Estimación de la Integral")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.show()
```

se desea calcular la esperanza de la vida útil de un satélite, cuyoequipamiento
principal (sujeto a fallos) consiste en dos computadoras y unequipamiento de transmisión. El
satélite funciona mientras al menos una de las dos computadoras y el equipo de transmisi ́on
funcionen. En el momento del lanzamiento, las computadoras poseen una vida útil aleatoriade
distribución uniforme entre 0 y 500 hs; y el equipamiento de transmisi ́onposee una vida útil
aleatoria de distribución uniforme entre 0 y 1500 hs.
Solución: Definir un experimento consistente en sortear la vida útil (el tiempo hasta falla) de
las dos computadoras y el equipo de transmisión, yen base a estos calcular el tiempo hasta falla
del satélite. Para sortear estos valores, se emplea una rutina que genera números
seudoaleatorios.
Repetir este experimento un número fijo de veces (por ejemplo 1000)

```python
import numpy as np

# Definir la función que simula la vida útil del satélite
def vida_util_satellite():
    # Vida útil de las computadoras (distribución uniforme entre 0 y 500)
    vida_computadora1 = np.random.uniform(0, 500)
    vida_computadora2 = np.random.uniform(0, 500)

    # Vida útil del equipo de transmisión (distribución uniforme entre 0 y 1500)
    vida_transmision = np.random.uniform(0, 1500)

    # El satélite falla cuando todos los componentes fallan (el mínimo de los 3)
    tiempo_falla = min(vida_computadora1, vida_computadora2, vida_transmision)

    return tiempo_falla

# Número de simulaciones de Monte Carlo
num_simulaciones = 1000

# Realizar las simulaciones y calcular la vida útil estimada del satélite
tiempos_falla = [vida_util_satellite() for _ in range(num_simulaciones)]

# Estimar la esperanza de vida del satélite como el promedio de las simulaciones
esperanza_vida = np.mean(tiempos_falla)

# Mostrar el resultado
print(f"Esperanza de vida útil del satélite: {esperanza_vida:.2f} horas")
```

-Random walk

```python
import numpy as np

# Función para realizar una caminata aleatoria en una dimensión
def caminata_aleatoria(num_pasos):
    posicion = 0
    for _ in range(num_pasos):
        # Elegimos al azar si el paso es hacia la izquierda (-1) o hacia la derecha (+1)
        paso = np.random.choice([-1, 1])
        posicion += paso
    return posicion

# Simulación de Monte Carlo para estimar la probabilidad de regresar al origen
def estimar_probabilidad_regreso(num_simulaciones, num_pasos):
    regresos = 0
    for _ in range(num_simulaciones):
        if caminata_aleatoria(num_pasos) == 0:
            regresos += 1
    return regresos / num_simulaciones

# Parámetros
num_simulaciones = 10000
num_pasos = 100

# Estimación de la probabilidad
probabilidad = estimar_probabilidad_regreso(num_simulaciones, num_pasos)

print(f"Probabilidad estimada de regresar al origen: {probabilidad}")
```

Estimación de área bajo la curva

```python
import numpy as np

# Definir la función exponencial f(x) = e^(-x)
def f(x):
    return np.exp(-x)

# Simulación de Monte Carlo para estimar el área bajo la curva
def estimar_area(funcion, a, b, num_puntos):
    # Generamos puntos aleatorios en el intervalo [a, b]
    x_random = np.random.uniform(a, b, num_puntos)

    # Evaluamos la función en esos puntos
    y_random = funcion(x_random)

    # El área estimada es el promedio de los valores de la función multiplicado por el largo del intervalo
    area_estimada = (b - a) * np.mean(y_random)
    return area_estimada

# Parámetros
a = 0  # Límite inferior
b = 5  # Límite superior
num_puntos = 100000  # Número de puntos aleatorios

# Estimación del área bajo la curva
area = estimar_area(f, a, b, num_puntos)

print(f"Área estimada bajo la curva e^(-x) en el intervalo [{a}, {b}]: {area}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

class MonteCarloEstimation:
    def __init__(self, a, b, n, func_type, params):
        """
        a, b: Intervalo [a, b] sobre el cual se calculará el área.
        n: Número de muestras a generar.
        func_type: Tipo de función a usar: 'polinomio', 'exponencial', 'trigonometrica'.
        params: Parámetros de la función. Por ejemplo, para el polinomio, [a0, a1, ..., an]
                y para la exponencial, [a, b] (a * exp(b * x)).
        """
        self.a = a
        self.b = b
        self.n = n
        self.func_type = func_type
        self.params = params
        self.x_values = None
        self.y_values = None
        self.area_estimate = None

    def generate_samples(self):
        """Generar muestras aleatorias uniformes dentro del intervalo [a, b]."""
        self.x_values = np.random.uniform(self.a, self.b, self.n)

    def evaluate_function(self):
        """Evaluar la función seleccionada en las muestras generadas."""
        if self.func_type == 'polinomio':
            self.y_values = np.polyval(self.params, self.x_values)  # Polinomio: f(x) = a0 + a1*x + ... + an*x^n
        elif self.func_type == 'exponencial':
            a, b = self.params
            self.y_values = a * np.exp(b * self.x_values)  # Exponencial: f(x) = a * exp(b * x)
        elif self.func_type == 'trigonometrica':
            a1, b1, a2, b2 = self.params
            self.y_values = a1 * np.sin(b1 * self.x_values) + a2 * np.cos(b2 * self.x_values)  # Trigonométrica: f(x) = a1*sin(b1*x) + a2*cos(b2*x)

    def estimate_area(self):
        """Estimar el área bajo la curva utilizando el algoritmo de Monte Carlo."""
        self.area_estimate = ((self.b - self.a) / self.n) * np.sum(self.y_values)

    def plot_results(self):
        """Graficar los resultados."""
        plt.plot(self.x_values, self.y_values, 'bo', label='Muestras')
        plt.title(f'Estimación del área bajo la curva usando Monte Carlo')
        plt.xlabel('x')
        plt.ylabel('f(x)')
        plt.grid(True)
        plt.show()

    def run(self):
        """Ejecutar todo el proceso: generación de muestras, evaluación de la función y estimación del área."""
        self.generate_samples()
        self.evaluate_function()
        self.estimate_area()
        self.plot_results()

        # Imprimir resultados
        print(f"\nEstimación del área bajo la curva: {self.area_estimate:.4f}")
        print(f"\nMuestras generadas (x): {self.x_values}")
        print(f"Valores de la función (f(x)): {self.y_values}")

def main():
    # Solicitar entradas al usuario
    print("Selecciona el tipo de función:")
    print("1. Polinomio de grado n (ej. a0 + a1*x + a2*x^2 + ... + an*x^n)")
    print("2. Función Exponencial (ej. a * exp(b * x))")
    print("3. Función Trigonométrica (ej. a1*sin(b1*x) + a2*cos(b2*x))")

    func_choice = int(input("Ingresa el número de la opción deseada (1-3): "))

    a = float(input("Ingrese el límite inferior del intervalo (a): "))
    b = float(input("Ingrese el límite superior del intervalo (b): "))
    n = int(input("Ingrese el número de muestras (n): "))

    if func_choice == 1:
        # Polinomio de grado n
        degree = int(input("Ingrese el grado del polinomio: "))
        params = [float(input(f"Ingrese el coeficiente a{i}: ")) for i in range(degree + 1)]
        func_type = 'polinomio'
    elif func_choice == 2:
        # Función Exponencial
        a_param = float(input("Ingrese el valor de 'a' en la función exponencial: "))
        b_param = float(input("Ingrese el valor de 'b' en la función exponencial: "))
        params = [a_param, b_param]
        func_type = 'exponencial'
    elif func_choice == 3:
        # Función Trigonométrica
        a1 = float(input("Ingrese el valor de 'a1' en la función trigonométrica: "))
        b1 = float(input("Ingrese el valor de 'b1' en la función trigonométrica: "))
        a2 = float(input("Ingrese el valor de 'a2' en la función trigonométrica: "))
        b2 = float(input("Ingrese el valor de 'b2' en la función trigonométrica: "))
        params = [a1, b1, a2, b2]
        func_type = 'trigonometrica'

    # Crear una instancia de MonteCarloEstimation y ejecutar el proceso
    montecarlo = MonteCarloEstimation(a, b, n, func_type, params)
    montecarlo.run()

if __name__ == "__main__":
    main()
```

#Algoritmos Pseudoaleatorios
- No congruencial: Método del cuadrado Medio

```python
import pandas as pd

def metodo_cuadrado_medio(semilla, digitos, numeros):
    # Inicializamos las listas que se usarán para el DataFrame
    numeros_generados = []
    numeros_cuadrados = []
    centros = []
    valores_1 = []
    valores_2 = []

    numero_actual = semilla

    numeros_generados.append(semilla)

    for i in range(numeros):
        cuadrado = numero_actual ** 2
        # Añadir cuadrado a la lista
        numeros_cuadrados.append(cuadrado)

        # Convertir el cuadrado a string para poder eliminar los dígitos de los extremos
        cuadrado_str = str(cuadrado)

        # Sacar la posición del centro
        longitud = len(cuadrado_str)
        inicio_central = (longitud - digitos) // 2
        numero_central = cuadrado_str[inicio_central:inicio_central + digitos]

        centros.append(numero_central)

        # Convertir a entero la parte central
        numero_actual = int(numero_central)

        # Si el cuadrado tiene más de 3 dígitos, dividimos en dos valores
        if len(numero_central) > 3:
            valor1 = int(numero_central[:3])
            valor2 = int(numero_central[-3:])
            valores_1.append(valor1)
            valores_2.append(valor2)
            numero_actual = valor1  # Para la siguiente iteración
        else:
            valor1 = int(numero_central)
            valor2 = 0
            valores_1.append(valor1)
            valores_2.append(valor2)
            numero_actual = valor1

        # Almacenamos el valor generado
        numeros_generados.append(numero_actual)

    #Eliminar el último valor
    numeros_generados.pop(numeros)

    # Crear el DataFrame al final, después del bucle
    df = pd.DataFrame({
        "R(n)": numeros_generados,
        "R(n)^2": numeros_cuadrados,
        "M.R(n)^2": centros,
        "Valor 1": valores_1,
        "Valor 2": valores_2
    })

    return df

semilla = int(input("Ingrese el número de semilla: "))
numeros = int(input("Ingrese la cantidad de números a generar: "))
digitos = 4 #Se extraen 4 dígitos

df = metodo_cuadrado_medio(semilla, digitos, numeros)
df
```

```python
import random
import time
import math
import pandas as pd

def caminata_browniana(dim, pasos):
    # Inicializar la posición de la partícula en el origen
    posicion = [0] * dim
    distancia_maxima = 0

    # Realizar el movimiento browniano
    for _ in range(pasos):
        # Elegir un eje aleatorio
        eje = random.randint(0, dim - 1)
        # Elegir si el movimiento es positivo o negativo
        movimiento = random.choice([-1, 1])
        # Actualizar la posición
        posicion[eje] += movimiento
        # Calcular la distancia desde el origen
        distancia = math.sqrt(sum([x**2 for x in posicion]))
        # Actualizar la distancia máxima si es necesario
        distancia_maxima = max(distancia_maxima, distancia)

    return distancia_maxima

def realizar_experimento(dim, pasos, repeticiones):
    distancias_maximas = []
    tiempos = []

    # Realizar el experimento 'repeticiones' veces
    for _ in range(repeticiones):
        # Medir el tiempo de ejecución de cada caminata
        inicio = time.time()
        distancia_maxima = caminata_browniana(dim, pasos)
        fin = time.time()

        # Almacenar los resultados
        distancias_maximas.append(distancia_maxima)
        tiempos.append(fin - inicio)

    # Calcular el promedio de la distancia máxima
    promedio_distancia_maxima = sum(distancias_maximas) / len(distancias_maximas)

    # Crear un dataframe para mostrar los resultados
    resultados = pd.DataFrame({
        'Distancia Maxima': distancias_maximas,
        'Tiempo (segundos)': tiempos
    })

    return resultados, promedio_distancia_maxima

def main():
    # Configuración del experimento
    dim = int(input("Ingrese la dimensión del sistema de coordenadas (1 o 2): "))
    pasos = int(input("Ingrese el número de pasos: "))
    repeticiones = int(input("Ingrese el número de repeticiones: "))

    # Realizar el experimento
    resultados, promedio_distancia_maxima = realizar_experimento(dim, pasos, repeticiones)

    # Mostrar los resultados
    print("\nResultados del experimento:")
    print(resultados)
    print(f"\nPromedio de la distancia máxima: {promedio_distancia_maxima:.4f}")
    print(f"Tiempo total de ejecución: {sum(resultados['Tiempo (segundos)']):.4f} segundos")

if __name__ == "__main__":
    main()
```

Congruencial lineal

```python
class CongruencialMixto:
    def __init__(self, a, c, m, semilla):
        self.a = a
        self.c = c
        self.m = m
        self.x = semilla

    def generar_numero(self):

        return self.x

    def calcular_intermedio(self):

        return self.a * self.x + self.c

    def aplicar_modulo(self):

        return (self.calcular_intermedio()) % self.m

    def generar_siguiente(self):

        self.x = self.aplicar_modulo()
        return self.x


def generar_congruencial(a, c, m, semilla, cantidad):

    generador = CongruencialMixto(a, c, m, semilla)
    resultados = []

    for n in range(0, cantidad + 1):
        x_n = generador.generar_numero()
        intermedio = generador.calcular_intermedio()
        modulo = generador.aplicar_modulo()
        resultados.append((n, x_n, intermedio, modulo))
        generador.generar_siguiente()

    return resultados


def main():
    try:
        a = int(input("Ingresa el valor de 'a' (Multiplicador): "))
        c = int(input("Ingresa el valor de 'c' (Incremento): "))
        m = int(input("Ingresa el valor de 'm' (Módulo): "))
        semilla = int(input("Ingresa el valor de la semilla: "))
        cantidad = int(input("Ingresa la cantidad de números a generar: "))

        if m <= 0 or a <= 0 or c < 0 or semilla < 0 or cantidad <= 0:
            raise ValueError("Los valores deben ser positivos.")

        resultados = generar_congruencial(a, c, m, semilla, cantidad)

        print("\nTabla de Resultados:")
        print(f"{'n':<5} {'X(n)':<10} {'a*X(n)+c':<15} {'[a*X(n)+c] mod m'}")
        print("-" * 50)
        for resultado in resultados:
            n, x_n, intermedio, modulo = resultado
            print(f"{n:<5} {x_n:<10} {intermedio:<15} {modulo}")

    except ValueError as e:
        print(f"Error: {e}")


if __name__ == "__main__":
    main()
```
